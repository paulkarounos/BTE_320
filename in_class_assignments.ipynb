```python
sales = float(input("Enter sales ($): "))
tax_rate = float(input("Enter tax rate (%): "))

cogs = 0.40 * sales
gross_profit = sales - cogs
taxes = gross_profit * (tax_rate / 100)
net_income = gross_profit - taxes

print(f"Sales: $ {sales:,.2f}")
print(f"Cost of Goods Sold: $ {cogs:,.2f}")
print(f"Gross Profit: $ {gross_profit:,.2f}")
print(f"Taxes: $ {taxes:,.2f}")
print(f"Net Income: $ {net_income:,.2f}")
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

if op == "+":
    result = num1 + num2
elif op == "-":
    result = num1 - num2
elif op == "*":
    result = num1 * num2
elif op == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error: Division by zero"
else:
    result = "Invalid operator"

print("Result:", result)
```

```python
print("Ad$ Profit")
for ad in range(0, 201, 25):
    attendees = 20 + 2 * (ad ** 0.5)
    revenue = attendees * 10
    profit = revenue - ad - 200
    print(float(ad), profit)
```

```python
import random

play = input("Do you want to play? (yes/no): ")
while play.lower() == "yes":
    user = input("Choose rock, paper, or scissors: ").lower()
    computer = random.choice(["rock", "paper", "scissors"])
    print("Computer chose:", computer)

    if user == computer:
        print("Tie!")
    elif (user == "rock" and computer == "scissors") or \
         (user == "paper" and computer == "rock") or \
         (user == "scissors" and computer == "paper"):
        print("You win!")
    else:
        print("Computer wins!")

    play = input("Play again? (yes/no): ")
```

```python
def reverse(s):
    if len(s) <= 1:
        return s
    return reverse(s[1:]) + s[0]

print(reverse("hello"))
print(reverse("12345"))
```

```python
def checkEmailAddress(email):
    if email.count("@") != 1:
        return False
    username, rest = email.split("@")
    if len(username) == 0:
        return False
    if rest.count(".") < 1:
        return False
    domain_parts = rest.split(".")
    if len(domain_parts) < 2:
        return False
    for part in domain_parts:
        if len(part) == 0:
            return False
    return True

email = input("Enter email address: ")
print(checkEmailAddress(email))
```

```python
def getValues(n):
    values = []
    while len(values) < n:
        try:
            v = float(input(f"Enter value {len(values) + 1}: "))
            if v < 0:
                print("Invalid: value cannot be negative. Try again.")
                continue
            values.append(v)
        except ValueError:
            print("Invalid: please enter a number. Try again.")
    return values


def calcAverage(alist):
    trimmed = alist[3:]
    if len(trimmed) == 0:
        return 0
    return sum(trimmed) / len(trimmed)


try:
    n = int(input("How many values will you enter? (must be > 3): "))
    if n <= 3:
        print("Invalid: n must be greater than 3.")
    else:
        weights = getValues(n)
        avg = calcAverage(weights)
        print("Average (excluding first 3):", avg)
except ValueError:
    print("Invalid input for n. Please enter an integer.")
```

```python
class Greeting:
    def __init__(self, first, last):
        self.first = first
        self.last = last

    def display(self, formal):
        if formal:
            print(f"{self.last}, {self.first}")
        else:
            print(self.first)

    def __str__(self):
        return f"{self.first} {self.last}"


first = input("Enter first name: ")
last = input("Enter last name: ")
choice = input("Display formally? (True/False): ")
formal = choice.strip().lower() == "true"

g = Greeting(first, last)
g.display(formal)
```

```python
class Student:
    def __init__(self, name, number, courses=None):
        self.name = name
        self.number = number
        self.courses = courses if courses is not None else []

    def getName(self):
        return self.name

    def getCourses(self):
        return self.courses

    def setName(self, name):
        self.name = name

    def __len__(self):
        return len(self.courses)

    def enroll(self, course):
        if course not in self.courses:
            self.courses.append(course)


name = input("Enter student name: ")
number = int(input("Enter student number: "))
s = Student(name, number)

course = input("Enter a course to enroll in (or blank to stop): ")
while course != "":
    s.enroll(course)
    course = input("Enter a course to enroll in (or blank to stop): ")

print("Name:", s.getName())
print("Courses:", s.getCourses())
print("Number of courses:", len(s))
```
